### Example 1 : A Simple Cost Function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x ** 2 + x + 1
```

#### Slope & Derivative
Create a python function that for the derivative of f(x) called dif(f(x))

```python
def df(x):
    return 2 * x + 1
```

```python
#make data
x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
x_1
```

### Python loops and gradient descent

```python
#Python loops example
for n in range(5):
    print("Hello World")
print("End of loop")
```

```python
# Gradient Descent
new_x = -10
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - gradient * step_multiplier
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    #Step size allow you to know when to break the loop of gradient descent, when gradient approaches 0, step_size will also approaches 0
    step_size = abs(new_x - previous_x)
    print(step_size)
    
    #Our step size is always getting smaller and smaller because the error is reducing
    if(step_size < precision):
        break
    
print("local minimum occurs at {}".format(new_x))
print("slope at this minimum point is {}".format(df(new_x)))
print("Cost or f(x) at this points is {}".format(f(new_x)))
print(x_list)
print(slope_list)
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,3,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_1, f(x_1))
plt.title("Cost function", fontsize=17)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("X")
plt.ylabel("f(x)")

values = np.array(x_list)
plt.scatter(x_list, f(values), color="red", s=100, alpha=0.5)

# Derivative of Cost Function
plt.subplot(1,3,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.plot(x_1, df(x_1), linewidth=3, alpha=0.4)

plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)

# Gradient Descent
plt.subplot(1,3,3) #(1,2,2): 1 row, 2 columns, Column no.2
plt.title("Gradient Descent", fontsize=17)
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.grid()
plt.plot(x_1, df(x_1))

plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)

plt.show()
```

## Example 2: Multiple Minima vs Initial Guess & Advanced Function
$ g(x) = x^4 - 4x^2 + 5$

```python
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x ** 4 - 4 * x ** 2 + 5

def dg(x):
    return 4 * x ** 3 - 8 * x 
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,2,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_2, g(x_2))
plt.title("Cost function", fontsize=17)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("X")
plt.ylabel("g(x)")


# Derivative of Cost Function
plt.subplot(1,2,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.plot(x_2, dg(x_2), linewidth=3, alpha=0.4)
```

## Gradient Descent as Python Function

```python
# Gradient Descent

def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter = 300):
    
    #Set initial_guess to new_x
    new_x = initial_guess

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - gradient * multiplier

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        #Step size allow you to know when to break the loop of gradient descent, when gradient approaches 0, step_size will also approaches 0
        step_size = abs(new_x - previous_x)

        #Our step size is always getting smaller and smaller because the error is reducing
        if(step_size < precision):
            break
            
    return new_x, x_list, slope_list #returning a tuple
```

```python
local_min, list_x, deriv_list = gradient_descent(dg,2,0.01,0.001 ) #sequential unpacking tuple
```

```python
print("Local min occurs at {}".format(local_min))
print("Number of steps {}".format(len(list_x)))
```

```python
print("Local min occurs at {}".format(local_min))
print("Number of steps {}".format(len(list_x)))
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,2,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_2, g(x_2))
plt.title("Cost function", fontsize=17)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("X")
plt.ylabel("g(x)")

plt.scatter(list_x, g(np.array(list_x)), color="red", s=100, alpha=0.5)


# Derivative of Cost Function
plt.subplot(1,2,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.plot(x_2, dg(x_2), linewidth=3, alpha=0.4)

plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.5)
```

## Example 4: Data Vizualization with 3D Chart
### Minimise $$f(x,y) = \frac{1}{3^{-x ^2 - y ^2 } + 1} $$
Minimise $$f(x,y)= \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
# define function
def f(x,y):
    r = 3**(-x**2 * -y**2)
    return 1/(r + 1)

# Make our x and y data
x_4 = np.linspace(start=-2, stop=2, num=200)
y_4 = np.linspace(start=-2, stop=2, num=200)

print("Shape of Array {}".format(x_4.shape))

x_4,y_4 = np.meshgrid(x_4,y_4) # returns 2-Dimensional Array representing the X and Y coordinates of all the points and thus allowing to plot a 3-D graph. Link for reference: https://www.geeksforgeeks.org/numpy-meshgrid-function/

print("Shape of Array {}".format(x_4.shape))
```

```python
# generating 3d plot
fig = plt.figure(figsize=[16,12]) # return a figure object
ax = fig.gca(projection='3d') # return a Axes3dSubplot, basically a 3d figure

ax.set_xlabel("X", fontsize=20)
ax.set_ylabel("Y", fontsize=20)
ax.set_zlabel("f(x,y) - Cost", fontsize=20)

ax.plot_surface(x_4, y_4, f(x_4,y_4), cmap=cm.coolwarm, alpha=0.4) #coolwarm, winter, hot
plt.show()
```
