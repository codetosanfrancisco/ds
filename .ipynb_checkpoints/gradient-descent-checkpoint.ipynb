### Example 1 : A Simple Cost Function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x ** 2 + x + 1
```

#### Slope & Derivative
Create a python function that for the derivative of f(x) called dif(f(x))

```python
def df(x):
    return 2 * x + 1
```

```python
#make data
x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
x_1
```

### Python loops and gradient descent

```python
#Python loops example
for n in range(5):
    print("Hello World")
print("End of loop")
```

```python
# Gradient Descent
new_x = -10
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - gradient * step_multiplier
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    #Step size allow you to know when to break the loop of gradient descent, when gradient approaches 0, step_size will also approaches 0
    step_size = abs(new_x - previous_x)
    print(step_size)
    
    #Our step size is always getting smaller and smaller because the error is reducing
    if(step_size < precision):
        break
    
print("local minimum occurs at {}".format(new_x))
print("slope at this minimum point is {}".format(df(new_x)))
print("Cost or f(x) at this points is {}".format(f(new_x)))
print(x_list)
print(slope_list)
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,3,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_1, f(x_1))
plt.title("Cost function", fontsize=17)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("X")
plt.ylabel("f(x)")

values = np.array(x_list)
plt.scatter(x_list, f(values), color="red", s=100, alpha=0.5)

# Derivative of Cost Function
plt.subplot(1,3,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.plot(x_1, df(x_1), linewidth=3, alpha=0.4)

plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)

# Gradient Descent
plt.subplot(1,3,3) #(1,2,2): 1 row, 2 columns, Column no.2
plt.title("Gradient Descent", fontsize=17)
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.grid()
plt.plot(x_1, df(x_1))

plt.scatter(x_list, slope_list, color="red", s=100, alpha=0.5)

plt.show()
```

## Example 2: Multiple Minima vs Initial Guess & Advanced Function
$ g(x) = x^4 - 4x^2 + 5$

```python
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x ** 4 - 4 * x ** 2 + 5

def dg(x):
    return 4 * x ** 3 - 8 * x 
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,2,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_2, g(x_2))
plt.title("Cost function", fontsize=17)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("X")
plt.ylabel("g(x)")


# Derivative of Cost Function
plt.subplot(1,2,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.plot(x_2, dg(x_2), linewidth=3, alpha=0.4)
```

## Gradient Descent as Python Function

```python
# Gradient Descent

def gradient_descent(derivative_func, initial_guess, multiplier, precision, max_iter = 300):
    
    #Set initial_guess to new_x
    new_x = initial_guess

    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - gradient * multiplier

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        #Step size allow you to know when to break the loop of gradient descent, when gradient approaches 0, step_size will also approaches 0
        step_size = abs(new_x - previous_x)

        #Our step size is always getting smaller and smaller because the error is reducing
        if(step_size < precision):
            break
            
    return new_x, x_list, slope_list #returning a tuple
```

```python
local_min, list_x, deriv_list = gradient_descent(dg,2,0.01,0.001 ) #sequential unpacking tuple
```

```python
print("Local min occurs at {}".format(local_min))
print("Number of steps {}".format(len(list_x)))
```

```python
local_min, list_x, deriv_list = gradient_descent(dg,0,0.01,0.001 )
```

```python
print("Local min occurs at {}".format(local_min))
print("Number of steps {}".format(len(list_x)))
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,2,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_2, g(x_2))
plt.title("Cost function", fontsize=17)
plt.xlim(-2,2)
plt.ylim(0.5,5.5)
plt.xlabel("X")
plt.ylabel("g(x)")

plt.scatter(list_x, g(np.array(list_x)), color="red", s=100, alpha=0.5)


# Derivative of Cost Function
plt.subplot(1,2,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.xlim(-2,2)
plt.ylim(-6,8)
plt.plot(x_2, dg(x_2), linewidth=3, alpha=0.4)

plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.5)
```

## Example 3: Divergence, Overflow & Python Tuples
$h(x) = x ^ 5 - 2x^4 + 2$

```python
x_3 = np.linspace(start=-2.5,stop=2.5 ,num=1000)

def h(x):
    return x ** 5 - 2 * x ** 4 + 2

def dh(x):
    return 5 * x ** 4 - 8 * x ** 3 
```

```python
local_min, list_x, deriv_list = gradient_descent(dh,-0.5,0.01,0.001,20)
```

```python
print("Local min occurs at {}".format(local_min))
print("Number of steps {}".format(len(list_x)))
print("Cost at minimum is {}".format(h(local_min)))
```

```python
# plot function and derivate side by side

plt.figure(figsize=[15,5])

# Cost function
plt.subplot(1,2,1) #(1,2,1) : 1 row, 2 columns, Column no.1
plt.plot(x_3, h(x_3))
plt.title("Cost function", fontsize=17)
plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.xlabel("X")
plt.ylabel("g(x)")

plt.scatter(list_x, h(np.array(list_x)), color="red", s=100, alpha=0.5)


# Derivative of Cost Function
plt.subplot(1,2,2) #(1,2,2): 1 row, 2 columns, Column no.2
plt.xlabel("X")
plt.ylabel("dg(x)")
plt.title("The Derivative of Cost function", fontsize=17)
plt.grid()
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.plot(x_3, dh(x_3), linewidth=3, alpha=0.4)

plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.5)
```
